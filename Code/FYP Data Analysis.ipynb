```python
import pandas as pd
import numpy as np
```

```python
# Load data from the CSV file
file_path = 'FYP Data Result.csv'  # Replace with your file path
data = pd.read_csv(file_path)
```

```python
data
```

```python
# Assuming your CSV has columns named 'Actual_Length' and 'Predicted_Length'
actual_lengths = data['Actual_Length'].values
predicted_lengths = data['Predicted_Length'].values

# Calculate Mean Absolute Error (MAE)
mae = np.mean(np.abs(actual_lengths - predicted_lengths))
print("Mean Absolute Error (MAE):", mae)

# Calculate Mean Squared Error (MSE)
mse = np.mean((actual_lengths - predicted_lengths) ** 2)
print("Mean Squared Error (MSE):", mse)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
# Assuming your CSV has columns named 'Actual_Growth' and 'Predicted_Growth'
actual_growths = data['Actual_Growth'].values
predicted_growths = data['Predicted_Growth'].values

# Calculate Mean Absolute Error (MAE)
mae = np.mean(np.abs(actual_growths - predicted_growths))
print("Mean Absolute Error (MAE):", mae)

# Calculate Mean Squared Error (MSE)
mse = np.mean((actual_growths - predicted_growths) ** 2)
print("Mean Squared Error (MSE):", mse)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_bar_graph(data, labels, title):
    fig, ax = plt.subplots()

    width = 0.25
    x = np.arange(len(labels))

    ax.bar(x - width, data['Abu']['Random Forest'], width, label='Random Forest - Abu')
    ax.bar(x - width, data['Abu']['Decision Tree'], width, label='Decision Tree - Abu')
    ax.bar(x - width, data['Abu']['KNN'], width, label='KNN - Abu')
    ax.bar(x - width, data['Abu']['SVM'], width, label='SVM - Abu')

    ax.bar(x + width, data['Bakar']['Random Forest'], width, label='Random Forest - Bakar')
    ax.bar(x + width, data['Bakar']['Decision Tree'], width, label='Decision Tree - Bakar')
    ax.bar(x + width, data['Bakar']['KNN'], width, label='KNN - Bakar')
    ax.bar(x + width, data['Bakar']['SVM'], width, label='SVM - Bakar')

    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

    ax.set_title(title)

    plt.tight_layout()
    plt.show()

data = {
    'Abu': {
        'Random Forest': 2,
        'Decision Tree': 4,
        'KNN': 6,
        'SVM': 8
    },
    'Bakar': {
        'Random Forest': 1,
        'Decision Tree': 3,
        'KNN': 5,
        'SVM': 7
    }
}

labels = ['Random Forest', 'Decision Tree', 'KNN', 'SVM']

plot_bar_graph(data, labels, 'Bar Graph with Two Labels (Abu and Bakar)')
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

# set width of bar 
barWidth = 0.25
fig, ax = plt.subplots(figsize=(12, 8)) 

# set height of bar 
IT = [0.4053, 0.2275, 0.2773, 0.1478] 
ECE = [0.8556, 0.8301, 0.5608, 0.5606] 

# Set position of bar on X axis 
br1 = np.arange(len(IT)) 
br2 = [x + barWidth for x in br1] 

# Make the plot
plt.bar(br1, IT, color ='r', width=barWidth, 
        edgecolor='grey', label='Case A') 
plt.bar(br2, ECE, color ='g', width=barWidth, 
        edgecolor='grey', label='Case B') 

# Adding data values on top of the bars
for i, v in enumerate(IT):
    ax.text(i, v + 0.02, str(v), color='black', ha='center', va='bottom', fontweight='bold')

for i, v in enumerate(ECE):
    ax.text(i + barWidth, v + 0.02, str(v), color='black', ha='center', va='bottom', fontweight='bold')

# Adding Xticks 
plt.xlabel('Model Score', fontweight='bold', fontsize=15) 
plt.ylabel('Value', fontweight='bold', fontsize=15) 
plt.xticks([r + barWidth/2 for r in range(len(IT))], 
           ['Random forest', 'Decision Tree', 'KNN', 'SVM'])

plt.legend()
plt.show()
```
